```python
from json import dumps,loads
from base64 import b64encode
```

```python
ENCODING = 'utf-8'
IMAGE_NAME = 'rubik.png'
JSON_NAME = 'output.json'
```

```python
# first: reading the binary stuff
# note the 'rb' flag
# result: bytes
with open(IMAGE_NAME, 'rb') as open_file:
    byte_content = open_file.read()
```

```python
# second: base64 encode read data
# result: bytes (again)
base64_bytes = b64encode(byte_content)
```

```python
# third: decode these bytes to text
# result: string (in utf-8)
base64_string = base64_bytes.decode(ENCODING)
```

```python
base64_bytes
```

```python
# optional: doing stuff with the data
# result here: some dict
raw_data = {IMAGE_NAME: base64_string}
```

```python
# now: encoding the data to json
# result: string
json_data = dumps(raw_data, indent=2)
```

```python
json_data
```

```python
# headers = {'Content-type': 'application/json; charset=utf-8'}
```

```python
headers = {'Content-type': 'image/jpg; charset=utf-8'}
```

```python
payload = json_data
```

```python
r = requests.post(url=URL, data=payload, headers = headers,params=PARAMS)
```

```python
r.headers
```

```python
r.status_code
```

```python
d=r.content
```

```python
d
```

```python
imgdata = base64.b64decode(d['rubik.png'])
filename = 'rubikReturned2.jpg'  # I assume you have a way of picking unique filenames
with open(filename, 'wb') as f:
    f.write(imgdata)
```
